```python
x = linspace(0.15,0.3,100)
y = linspace(0,0.5,100)

# make a 2D meshgrid in the transverse plane
cs2, Lambda_max = meshgrid(x,y)
Pi = (5/3+2*cs2)*Lambda_max-1

# make the contour plot
fig = plt.figure()
cont = plt.contourf(cs2, Lambda_max, Pi, levels, cmap="bwr", extend='both')
cbar = fig.colorbar(cont)
plt.xlabel(r"$c_s^2$")
plt.ylabel(r"$\Lambda_{max}$")
plt.xlim([0.15, 0.3])
plt.ylim([0, 0.6])
plt.tight_layout()
```
